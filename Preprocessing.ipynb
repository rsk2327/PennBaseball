```python
def processFile(filename, gameID):
    
    df= pd.read_excel(filename)
    
    dropColList =['\n','No', 'Col15', 'Col17', 'Col18', 'Col19', 'Col21',
     'Col23', 'Col24', 'Col25', 'Col26', 'Col27', 'Notes',
     'Supervisor', 'Col50', 'Col51', 'Col52', 'Col53', 'Col55', 'Col56', 'Col71', 'Col72',
     'Col73', 'Col74', 'Supervisor.1']
    
    for col in dropColList:
        if col in df.columns:
            df.drop(col,inplace = True, axis = 1)
    
    colMap = {}

    for col in list(df.columns):
        if len(col.split(" "))>1:
            terms = col.split(" ")
            terms = [x for x in terms if not any( t in x for t in ['[',']'] ) ]

            terms = [x.replace("(","").replace(")","") for x in terms]

            terms = [x for x in terms if x != "-"]
            colMap[col] = "_".join(terms)


    df = df.rename(colMap, axis=1)
    
    
    df['hit'] = df.Hit_Ball_Flight_Time.apply(lambda x : 0 if pd.isna(x) else 1)
    
    df['gameID'] = gameID
    
    return df
    
```

```python
dataFolder = './FlightScope_Data/'
files = [x for x in os.listdir(dataFolder) if 'xlsx' in x]
```

```python
files
```

```python
df = pd.DataFrame()
ctr = 0

for filename in files:
    
    temp = processFile(os.path.join(dataFolder, filename), gameID = ctr)
    ctr += 1
    df = df.append(temp)
    
```

```python
def cleanName(x):
    
    x = x.split(" ")
    x = [i for i in x if "(" not in i]
    
    x = " ".join(x)
    x = x.strip()
    
    return x
```

```python
df.Pitcher = df.Pitcher.apply(lambda x : cleanName(x))
```

```python
df.Batter = df.Batter.apply(lambda x : cleanName(x))
```
